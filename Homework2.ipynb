## Import Variables

```python
from Array import *
import random
import timeit
import matplotlib.pyplot as plt
import numpy as np
```

## Create array of a specified size with random items and insert those items

```python
def initArray (size = 100, maxVal = 100, seed = 3.14159):
    arr = Array(size)
    random.seed(seed)
    
    for i in range(size):
        arr.insert(random.randrange(maxVal))
    return arr
```

## Display random values

```python
arr = initArray()
print("Array containing", arr)
```

## Create a plot for the time it take to sort the random array

```python
sort_times = []
sort_algorithm = ['Bubble Sort', 'Selection Sort', 'Insertion Sort']
for test in ['initArray().bubbleSort()',
             'initArray().selectionSort()',
             'initArray().insertionSort()']:
    elapsed = timeit.timeit(test, number=100, globals=globals())
    sort_times.append(elapsed)
    print(test, "took", elapsed, "seconds", flush=True)

fig = plt.figure(figsize = (7,3.5))
ax = fig.add_subplot(111)
ax.set_title("Sorting Times")

ax.bar(sort_algorithm, sort_times)
ax.set_xlabel("Sort Functions")
ax.set_ylabel("Sorting Times")



```
